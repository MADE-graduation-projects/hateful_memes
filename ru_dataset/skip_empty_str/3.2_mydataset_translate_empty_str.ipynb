```python
data_dir = r'C:\Users\notebook\Downloads\_my_dataset\20221214_hateful_memes\hateful_memes\\'

train_path = data_dir + 'train_trans_es.jsonl'
dev_path = data_dir + 'test_trans_es.jsonl'

train_data = pd.read_json(train_path, lines=True)
test_data = pd.read_json(dev_path, lines=True)
```

```python
mname = "facebook/wmt19-ru-en"
tokenizer = FSMTTokenizer.from_pretrained(mname)
model = FSMTForConditionalGeneration.from_pretrained(mname)
```

```python
def translate(input):
    if input == '':
        return ''
    input_ids = tokenizer.encode(input, return_tensors="pt")
    outputs = model.generate(input_ids)
    decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return decoded
```

```python
translate("Машинное обучение - это здорово, не так ли?")
```

```python
for i, inp in enumerate(test_data['recognized_text_easyocr'].values):
    print(inp)
    print(translate(inp))
    print('\n')
    if i > 5:
        break
```

```python
dev_rows = pd.read_json(dev_path, lines=True)
with open(data_dir + 'test_trans_es.jsonl', "w", encoding="utf-8") as rf:
    for i, (_, row) in enumerate(dev_rows.iterrows()):
        dictRow = dict(row)
        dictRow["translated_text"] = translate(dictRow['recognized_text_easyocr'])
        rf.write(json.dumps(dictRow) + "\n")
```

```python
train_rows = pd.read_json(data_dir + 'train_ocr.jsonl', lines=True)
with open(data_dir + 'train_trans_es.jsonl', "w", encoding="utf-8") as rf:
    for i, (_, row) in enumerate(tqdm(train_rows.iterrows())):
        dictRow = dict(row)
        dictRow["translated_text"] = translate(dictRow['recognized_text_easyocr'])
        rf.write(json.dumps(dictRow) + "\n")
```
